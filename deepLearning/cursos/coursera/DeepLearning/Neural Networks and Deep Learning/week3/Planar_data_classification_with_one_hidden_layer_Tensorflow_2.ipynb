# 4. Neural network
**Here is our model**:
classification_kiank.png
**Mathematically**:

For one example $x^{(i)}$:
$$z^{[1] (i)} =  W^{[1]} x^{(i)} + b^{[1]}\tag{1}$$ 
$$a^{[1] (i)} = \tanh(z^{[1] (i)})\tag{2}$$
$$z^{[2] (i)} = W^{[2]} a^{[1] (i)} + b^{[2]}\tag{3}$$
$$\hat{y}^{(i)} = a^{[2] (i)} = \sigma(z^{ [2] (i)})\tag{4}$$
$$y^{(i)}_{prediction} = \begin{cases} 1 & \mbox{if } a^{[2](i)} > 0.5 \\ 0 & \mbox{otherwise } \end{cases}\tag{5}$$

Given the predictions on all the examples, you can also compute the cost $J$ as follows: 
$$J = - \frac{1}{m} \sum\limits_{i = 0}^{m} \large\left(\small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right)  \large  \right) \small \tag{6}$$

**Reminder**: The general methodology to build a Neural Network is to:
    1. Define the neural network structure ( # of input units,  # of hidden units, etc). 
    2. Initialize the model's parameters
    3. Loop:
        - Implement forward propagation
        - Compute loss
        - Implement backward propagation to get the gradients
        - Update parameters (gradient descent)

## 4.1. Create dense layer

```
class Dense(tf.Module):
  def __init__(self, in_features, out_features, activation='tanh', name=None):
    super().__init__(name=name)

    # Weight initialized with standard normal distribution 
    self.w = tf.Variable(tf.random.normal([in_features, out_features]), name='w')

    # bias initialized zeros
    self.b = tf.Variable(tf.zeros([out_features, 1]), name='b')

    # Activation function
    self.activation = activation

  # Calculate layer function
  def __call__(self, x):
    z = tf.tensordot(tf.transpose(self.w), x, axes = 1) + self.b
    if self.activation == 'tanh':
      a = tf.tanh(z)
    else:
      a = tf.sigmoid(z)
    return a
```

## 4.2. Create model

```
class MyModel(tf.Module):
  def __init__(self, **kwargs):
    super().__init__(**kwargs)

    # Implement neural network layers
    tf.random.set_seed(3)
    self.dense_1 = Dense(in_features = 2, out_features = 4)
    self.dense_2 = Dense(in_features = 4, out_features = 1, activation = 'sigmoid')

  # Calculate forward propagation
  def __call__(self, x):
    x = self.dense_1(x)
    return self.dense_2(x)
```

## 4.3. Create loss function and compute cost function


$$J = - \frac{1}{m} \sum\limits_{i = 1}^{m} \large{(} \small y^{(i)}\log\left(a^{[2] (i)}\right) + (1-y^{(i)})\log\left(1- a^{[2] (i)}\right) \large{)} \small\tag{13}$$

```
def cost(target_y, predicted_y):
  # Create loss function
  log = tf.math.multiply(target_y, tf.math.log(predicted_y))
  log_minus = tf.math.multiply(tf.subtract(1, target_y), tf.math.log(tf.subtract(1, predicted_y)))
  logprobs = tf.math.add(log, log_minus)

  # Compute cost funcion
  m = tf.constant(target_y.shape[1], dtype=tf.float32)
  cost = -1/m * tf.math.reduce_sum(logprobs)

  return cost
```

## 4.4. Implement training function

```
def train(model, x, y, learning_rate = 1.2):
  with tf.GradientTape() as t:
    current_cost = cost(y, model(x))
  
  # Calculate gradient (Backpropagation)
  grad = t.gradient(current_cost, model.trainable_variables)
  return grad, model.trainable_variables
```

```
# Define a training loop
def training_loop(model, x, y, epochs, learning_rate = 1.2):

  for epoch in range(1, epochs + 1):
    # Update the model with the single giant batch
    grad, trainable_variables = train(model, x, y, learning_rate)

    # Update parameters
    trainable_variables[0].assign_sub(grad[0] * learning_rate)
    trainable_variables[1].assign_sub(grad[1] * learning_rate)
    trainable_variables[2].assign_sub(grad[2] * learning_rate)
    trainable_variables[3].assign_sub(grad[3] * learning_rate)

    # Track this before I update
    current_cost = cost(y, model(x))

    if epoch == 1 or epoch % 1000 == 0:
      print('Epoch:', str(epoch), ', cost:', str(current_cost.numpy()))
```

## 4.5. Create predict function

```
def predict(predicted_y):
  predictions = tf.map_fn(lambda x: tf.cast(x > 0.5, dtype=tf.float32), predicted_y)
  return predictions
```

## 4.6. Create accuracy function

```
# Print accuracy
def get_accuracy(target_y, predictions):
  a = tf.tensordot(target_y, tf.transpose(predictions), axes = 1)
  b = tf.tensordot(1 - target_y, 1 - tf.transpose(predictions), axes = 1)
  print('Accuracy:', tf.squeeze(((a + b) / tf.constant(target_y.shape[1], dtype=tf.float32)) * 100).numpy(), '%')
  #print ('Accuracy: %d' % float((np.dot(Y.numpy(), predictions.numpy().T) + np.dot(1 - Y.numpy(), 1 - predictions.numpy().T))/float(Y.numpy().size)*100) + '%')
```
